## **Praktikum 1 : Implementasi Klasifikasi Naive Bayes Support Vector Machine (SVM)**

#### 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

#### 2. Import Data

```python
import pandas as pd
df = pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")
df.head(10)
```

#### 3. Data Analysis / Preprocessing

```python
df.head(10)
```

```python
df.isna().sum()
```

Diketahui bahwa tidak adanya data yang hilang pada dataset.

- Deskripsi Data

```python
df.describe()
```

Menggunakan fungsi describe untuk menampilkan deskripsi statistik dari data. Hal ini dapat menampilkan informasi statistik dari data yang kita punya, seperti menampilkan total data, rata-rata, standar deviasi, dll.

- Distribusi Data

```python
sns.histplot(df['age'], kde = True)
```

Menampilkan visualisasi berupa histogram berdasarkan variabel EstimatedSalary, untuk mengetahui persebaran data berdasarkan perkiraan gaji di dataset ini.

- Correlation Matrix

```python
df.corr()
```

```python
plt.figure()
sns.heatmap(df.corr())
plt.title('Plot Correlation Matrix', size=20)
plt.tight_layout()
plt.show()
```

#### 4. Split data

a). Variabel Independen / Dependen

```python
#split data into dependent/independent variables
X = df[['age', 'platelets']].values
y = df.iloc[:, -1].values
```

b). Train / Test Split

```python
#Split data into test/train set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = True)
```

#### 5. Feature Scaling

```python
# Scale dataset
sc = StandardScaler ()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#### 6. Train model

```python
#Classifier
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

#### 7. Predict Result / Score Model

```python
# Prediksi
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape( len (y_pred), 1), y_test.reshape( len (y_test),1)),1))
```

#### 8. Evaluasi Model

a). akurasi

```python
#Accuracy
accuracy_score(y_test, y_pred)
```

b) Classification Report

```python
# Classification report
print(f'Classification Report: \n{classification_report(y_test, y_pred)}')
```

c). F1 Score

```python
#F1 score
print(f"F score: {f1_score(y_test, y_pred)}")
```

d). Confusion Matrix

```python
cf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(cf_matrix, annot= True, fmt="d", cmap="Blues", cbar=False)
```

e). Precision-Recall curve

```python
# Plot Precision-Recall Curve
y_pred_proba = classifier.predict_proba(X_test) [:,1]
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_proba)

fig, ax = plt.subplots(figsize = (6,6))
ax.plot(recall, precision, label='Naive Bayes Classification', color = 'firebrick')
ax.set_title('Precision-Recall Curve')
ax.set_xlabel('Recall')
ax.set_ylabel('Precision')
plt.box (False)
ax.legend();
```

f). AUC/ROC curve

```python
# Plot AUC/ROC curve
y_pred_proba = classifier.predict_proba(X_test) [:,1]
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_proba)

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(fpr, tpr, label='Naive Bayes Classification', color = 'firebrick')
ax.set_title('ROC Curve')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
plt.box(False)
ax.legend();
```

#### 9. Visualization Predictions

- Prediction results on the training set

```python
# Visualize prediction results on training set

X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate (np.unique(y_set)):
  plt.scatter(X_set[y_set == j, 0], X_set[y_set==j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Naive Bayes (Training set)')
plt.xlabel('age')
plt.ylabel('platelets')
plt.legend()
plt.show()
```

- Prediction results on the test set

```python
# Visualize prediction results on test set

X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate (np.unique(y_set)):
  plt.scatter(X_set[y_set == j, 0], X_set[y_set==j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Naive Bayes (Test set)')
plt.xlabel('age')
plt.ylabel('platelets')
plt.legend()
plt.show()
```

- Example

```python
#Predict purchase with age (65) and platelets(327000)
print(classifier.predict(sc.transform([[65, 327000]])))
```
